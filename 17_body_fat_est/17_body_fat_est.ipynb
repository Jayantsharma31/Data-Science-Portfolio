```python
def drawplots(df,col):
    plt.figure(figsize=(15,7))
    plt.subplot(1,3,1)  #(1,3,1)--> 1 is basically the row and 3 is basically the columns and for 1st column i want histogram
    plt.hist(x=df[col],color='magenta')
    
    plt.subplot(1,3,2)   #(1,3,2)--> 1 is basically the row and 3 is basically the columns and for 2nd column i want probplot
    stats.probplot(x=df[col],dist='norm',plot=plt)   #probplot will help in find the distribution is gaussian or not
    plt.subplot(1,3,3)   #(1,3,3)--> 1 is basically the row and 3 is basically the columns and for 3rd column i want boxplot
    sns.boxplot(x=df[col],color='magenta')
    
    plt.show()  #it will draw all the columns
    
cols=list(df.columns)

for i in range(len(cols)):
    print(f'dsitribution plots for the feature {cols[i]} are shown below')
    drawplots(df,cols[i])
    
    print('#'*100)
```

```python
def correlation(df,threshold):   #passing df and threshold in the def function 
    colcor=set()  #we are using in order to remove duplicates,list allows duplicates but set doesnt allow 
    cormat=df.corr()   #gives correlation matrix
    for i in range(len(cormat)):  #iterating through rows
        for j in range(i):   #for every column in a row
            
            '''
            for each cell get the value of that cell by .iloc[i][j],
            where i is the row and j is the column if that abd(value) is greater
            than the threshold,get col_name and add it in the set
            '''
            
            
            
            if abs(cormat.iloc[i][j])>threshold:
                colname=cormat.columns[i]
                colcor.add(colname)
    return colcor
ans=correlation(train,threshold=0.85)
ans
```

**from the above feature selection techniques we can say that the features recomended by the Extra Trees Regressor and the mutual_information_gain are correct and from the correlation map we get to observe the similar pattern ,we noticed that abdomen and hip are having similar features .they're having collinearity,same goes with knee and thigh,we can either keep any one of them and we noticed that the feature abdomen gave more feature importance score to hip,so i will be selecting that**

# hyperparameter tuning

```python
#here it is params dictionary which is dict inside dict
params={
    'RandomForest':{
        'model':RandomForestRegressor(),
        'params':{
            'n_estimators':[int(x) for x in np.linspace(start=1,stop=1200,num=10)],
            'criterion':['mse','mae'],
            'max_depth':[int(x) for x in np.linspace(start=1,stop=30,num=5)],
            'min_samples_split':[2,5,10,12],
            'min_samples_leaf':[2,5,10,12],
            'max_features':[1,2,2,5,3,3,5,4,5],
        }
    },
    'D-tree':{
        'model':DecisionTreeRegressor(),
        'params':{
            'criterion':['mse','mae'],
            'splitter':['best','random'],
            'min_samples_split':[1,2,5,10,12],
            'min_samples_leaf':[1,2,5,10,12],
            'max_features':['auto','sqrt'],
            'ccp_alpha':[1,2,2,5,3,3,5,4,5],
        }
    },
    'SVM':{
        'model':SVR(),
        'params':{
            'C':[0.25,0.50,0.75,1.0],
            'tol':[1e-10,1e-5,1e-4,0.025,0.50,0.75],
            'kernel':['linear','poly','rbf','sigmoid'],
            'max_iter':[int(x) for x in np.linspace(start=1,stop=250,num=10)],
        }
    }
}
```

```python
scores=[]
for modelname, mp in params.items():
    clf=RandomizedSearchCV(mp['model'],param_distributions=mp['params'],
                           cv=5,n_jobs=-1,n_iter=10,scoring='neg_mean_squared_error')  #it will run 10*5 times=50--->50*3(bec we use params of 3 algos above)=150
    
    clf.fit(X_train,y_train)
    scores.append({
        'model_name':modelname,
        'best_score':clf.best_score_,
        'best_estimator':clf.best_estimator_
    })
#randomizedsearchcv will execute 10*5=50 times
# and we have 3 algorithms objects so the entire thing will run 150 times
#for regression we use neg_mean_squared_error as scoring
#if it was classification we would use accuracy
```

```python
scoresdf=pd.DataFrame(scores,columns=['model_name','best_score','best_estimator'])
scoresdf
```
